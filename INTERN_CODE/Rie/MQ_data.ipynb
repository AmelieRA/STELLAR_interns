```python
plt.figure(figsize = (10,8))
plt.scatter(table1['Teff'], table1['Prot'], s=0.1, c='r', label= 'Periodic')
plt.scatter(table2['Teff'], table2['Prot'], s=0.1, c='b', label = 'Non Periodic')
plt.xlim(6500, 3500)
plt.xlabel('Teff', fontsize = 20)
plt.ylabel('Rotation Period', fontsize = 20)
plt.title('Stellar Rotation Periods', fontsize = 30)
plt.legend(loc="upper right")
#title
#xlabel
#ylabel
#change font sizes
#plt legend google


```

```python
#create dt classifier
table1 = table1.dropna()
table2 = table2.dropna()
len(table1), len(table2)
```

```python
#assignment: combine data tables
#assignment: it is either going to belong to table1 or table2 (periodic good, non periodic bad)
#good we want 1, bad we want 0
#we want to create a column 
#row will be called our class
```

```python
#use len to be more accurate
valid = np.repeat(1, len(table1))
invalid = np.repeat(0, len(table2))
#use numpy append
classes = np.append(valid, invalid)
classes
```

```python
len(classes)
```

```python
#np.arange(0,10,1)
#start, end, interval
```

```python
tables = pd.concat([table1, table2], axis=0, sort=True)
#tables = tables.drop(columns=['Flag', 'Rper'])

tables   #, type(tables)
```

```python
tables
```

```python
tables['Class'] = classes
tables
```

```python
plt.figure(figsize = (10,8))
plt.scatter(tables['Teff'], tables['Prot'], s=0.1, c=tables['Class'], vmin=0,vmax=1)
plt.colorbar().set_label('Class',fontsize=20)
plt.xlabel('Teff', fontsize = 20)
plt.ylabel('Rotation Period', fontsize = 20)
plt.xlim(6500,3000);
plt.title('Stellar Rotation Periods', fontsize = 30)

```

```python
tables
```

```python
tables = tables.drop(columns=['Class', 'Flag', 'Rper'])
```

```python
tables
```

```python
#is this a reliable rotation period measurement

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = tables
Y = classes
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, random_state = 11)
```

```python
model = DecisionTreeClassifier()
#fit the model to the training set
fit_model = model.fit(train_X,train_Y)
#compare predictions from this model to the test set
predicted = fit_model.predict(test_X)
accuracy_score(test_Y, predicted)
```

```python
from sklearn import tree
tree.plot_tree(model.fit(test_X, test_Y)); 
#its w bc w is how believable it is
```
