```python
our_tics = []
our_temps = []
our_gaiabp = []
our_gaiarp = []
for row in ourdata1:
    tic = row['ID']
    temp = row['Teff']
    bp = row['gaiabp']
    rp = row['gaiarp']
    ourtics = ourdata2['tic'].values
    if tic in ourtics:
        our_tics.append(tic)
        our_temps.append(temp)
        our_gaiabp.append(bp)
        our_gaiarp.append(rp)
        #print(count)
        #count += 1
        ######apend all four to empty lists
    else:
        pass
    
    
    #bp and rp are colors ?
```
