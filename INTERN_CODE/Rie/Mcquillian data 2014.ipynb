```python
%matplotlib inline

import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
mq = pd.read_csv('data/McQuillan_data.csv')
mq
```

```python
mq #solar masses, mass compared to the sun

mass = mq['Mass']
rp = mq['Prot']
temp = mq['Teff']
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(temp, rp, c=mass, s=1)
plt.colorbar().set_label(label='Mass', size=20)
plt.xlim(6500,3000)
plt.xlabel("temperature in kelvin", size=20)
plt.ylabel("rotation period", size=20)
plt.title("McQuillan 2014 data", size=25)
```
